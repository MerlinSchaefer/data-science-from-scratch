## Visualizing Data

### Simple plot using matplotlib.pyplot

```python
from matplotlib import pyplot as plt
%matplotlib inline
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(years,gdp, color = "green", marker = "o", linestyle = "solid")
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.show()
plt.savefig('viz_gdp.png')
```

### Using the bar chart to create a histogram

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

#create buckets for grades by decile (100 with 90s)
histogram = Counter(min(grade//10*10,90)for grade in grades)

plt.bar([x+5 for x in histogram.keys()], #shift bars to right by 5
       histogram.values(), # set bar height to count of values
        10, #width of the bar
        edgecolor = (0,0,0),
       color = "gold") #black edge
plt.axis([-5,105,0,5]) #x-axis from -5-105 , y-axis from 0-5
plt.xticks([10*i for i in range(11)]) # tick every 10 numbers
plt.xlabel("Decile")
plt.ylabel("Number of Students")
plt.title("Distribution of Exam Grades")
plt.show()
plt.savefig("histogram.png")
```

### Importance of setting the x-axis in bar charts


```python
mentions  = [500,505]
years = [2017,2018]
plt.bar(years,mentions, 0.8)
plt.xticks(years)
#plt.ticklabel_format(useOffset = False)
plt.axis([2016.5,2018.5,499,506]) # bad
plt.show()
```

```python
plt.bar(years,mentions, 0.8)
plt.xticks(years)
#plt.ticklabel_format(useOffset = False)
plt.axis([2016.5,2018.5,0,550]) # better
plt.show()
```

```python
variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error  = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, "g-",label = "variance")
plt.plot(xs, bias_squared, "r-.", label = "bias squared")
plt.plot(xs, total_error, "b:", label = "total error")
plt.legend(loc = 9) # loc = 9 top middle
plt.xlabel("Model complexity")
plt.xticks([])
plt.title("Bias/Variance Trad-off")
plt.show()
plt.savefig("linechart.png")
```

### Scatterplot

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)
#label each dot
for label, friend_count, minute_count in zip(labels,friends,minutes):
    plt.annotate(label,
                xy = (friend_count,minute_count), #label with each point
                 xytext = (5,-5), # offset label slightly
                 textcoords = "offset points")

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("Daily Minutes")
plt.ylabel("Number of friends")
plt.show()
plt.savefig("scatter.png")
```

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```
